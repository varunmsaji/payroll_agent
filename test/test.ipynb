```python
import requests

BASE_URL = "http://127.0.0.1:8000/admin"   # change if needed


def test_get(url, params=None):
    print(f"\n‚û° GET {url}")
    r = requests.get(url, params=params)
    print("Status:", r.status_code)
    try:
        print("Response:", r.json())
    except:
        print("Response:", r.text)
    print("-" * 50)


# ------------------------------------------------------
# 1Ô∏è‚É£ Total Employees + Active Employees
# ------------------------------------------------------
test_get(f"{BASE_URL}/employees-summary")

# ------------------------------------------------------
# 2Ô∏è‚É£ Today's Attendance Summary
# ------------------------------------------------------
test_get(f"{BASE_URL}/attendance-today")

# ------------------------------------------------------
# 3Ô∏è‚É£ Late / On-time / Absent
# ------------------------------------------------------
test_get(f"{BASE_URL}/late-on-time-absent")

# ------------------------------------------------------
# 4Ô∏è‚É£ Overtime Today
# ------------------------------------------------------
test_get(f"{BASE_URL}/overtime-today")

# ------------------------------------------------------
# 5Ô∏è‚É£ Monthly Payroll Summary
# ------------------------------------------------------
month = 1
year = 2025
test_get(f"{BASE_URL}/payroll-summary/{month}/{year}")

# ------------------------------------------------------
# 6Ô∏è‚É£ Monthly Working Hours
# ------------------------------------------------------
test_get(f"{BASE_URL}/monthly-hours/{month}/{year}")

# ------------------------------------------------------
# 7Ô∏è‚É£ Recent Attendance Events
# ------------------------------------------------------
test_get(f"{BASE_URL}/recent-events")

# ------------------------------------------------------
# 8Ô∏è‚É£ Shift-wise Employee Distribution
# ------------------------------------------------------
test_get(f"{BASE_URL}/shift-distribution")
```

```python
import requests
import sys

BASE_URL = "http://127.0.0.1:8000"   # ‚úÖ Change only if your server runs elsewhere
EMPLOYEE_ID = 32                    # ‚úÖ Change to a real existing employee
MANAGER_ID = 27                     # ‚úÖ Change to real manager ID


# -----------------------------
# Helpers
# -----------------------------
def ok(msg):
    print(f"‚úÖ {msg}")

def fail(msg):
    print(f"\n‚ùå TEST FAILED: {msg}")
    sys.exit(1)

def check(r, step):
    if r.status_code != 200:
        fail(f"{step} failed ‚Üí {r.status_code} ‚Üí {r.text}")
    ok(step)


# -----------------------------
# 1Ô∏è‚É£ GET SINGLE EMPLOYEE
# -----------------------------
print("\n--- 1Ô∏è‚É£ Testing Employee Profile ---")
r = requests.get(f"{BASE_URL}/hrms/employee/{EMPLOYEE_ID}")
check(r, "Employee profile fetch")
print(r.json())


# -----------------------------
# 2Ô∏è‚É£ GET ALL EMPLOYEES (PAGINATION)
# -----------------------------
print("\n--- 2Ô∏è‚É£ Testing Employee List (Pagination) ---")
r = requests.get(f"{BASE_URL}/hrms/employees?page=1&limit=10")
check(r, "Employee list")
print(r.json())


# -----------------------------
# 3Ô∏è‚É£ GET ALL MANAGERS
# -----------------------------
print("\n--- 3Ô∏è‚É£ Testing Manager List ---")
r = requests.get(f"{BASE_URL}/hrms/employees/managers")
check(r, "Managers list")
print(r.json())


# -----------------------------
# 4Ô∏è‚É£ ASSIGN MANAGER
# -----------------------------
print("\n--- 4Ô∏è‚É£ Testing Assign Manager ---")
payload = {"manager_id": MANAGER_ID}
r = requests.put(
    f"{BASE_URL}/hrms/employee/{EMPLOYEE_ID}/manager",
    json=payload
)
check(r, "Manager assignment")
print(r.json())


# -----------------------------
# 5Ô∏è‚É£ CURRENT SHIFT
# -----------------------------
print("\n--- 5Ô∏è‚É£ Testing Current Shift ---")
r = requests.get(f"{BASE_URL}/hrms/employee/{EMPLOYEE_ID}/shift")
check(r, "Current shift")
print(r.json())


# -----------------------------
# 6Ô∏è‚É£ ATTENDANCE SUMMARY
# -----------------------------
print("\n--- 6Ô∏è‚É£ Testing Attendance Summary ---")
r = requests.get(f"{BASE_URL}/hrms/employee/{EMPLOYEE_ID}/attendance-summary")
check(r, "Attendance summary")
print(r.json())


# -----------------------------
# 7Ô∏è‚É£ TIME SUMMARY (Late + Overtime)
# -----------------------------
print("\n--- 7Ô∏è‚É£ Testing Time Summary ---")
r = requests.get(f"{BASE_URL}/hrms/employee/{EMPLOYEE_ID}/time-summary")
check(r, "Time summary")
print(r.json())


# -----------------------------
# 8Ô∏è‚É£ RECENT ATTENDANCE EVENTS
# -----------------------------
print("\n--- 8Ô∏è‚É£ Testing Recent Attendance Events ---")
r = requests.get(f"{BASE_URL}/hrms/employee/{EMPLOYEE_ID}/events")
check(r, "Attendance events")
print(r.json())


# -----------------------------
# 9Ô∏è‚É£ SALARY STRUCTURE
# -----------------------------
print("\n--- 9Ô∏è‚É£ Testing Salary Structure ---")
r = requests.get(f"{BASE_URL}/hrms/employee/{EMPLOYEE_ID}/salary")
check(r, "Salary structure")
print(r.json())


# -----------------------------
# üîü LATEST PAYROLL
# -----------------------------
print("\n--- üîü Testing Latest Payroll ---")
r = requests.get(f"{BASE_URL}/hrms/employee/{EMPLOYEE_ID}/payroll/latest")
check(r, "Latest payroll")
print(r.json())


# -----------------------------
# 1Ô∏è‚É£1Ô∏è‚É£ PAYROLL HISTORY
# -----------------------------
print("\n--- 1Ô∏è‚É£1Ô∏è‚É£ Testing Payroll History ---")
r = requests.get(f"{BASE_URL}/hrms/employee/{EMPLOYEE_ID}/payroll-history")
check(r, "Payroll history")
print(r.json())


# -----------------------------
# 1Ô∏è‚É£2Ô∏è‚É£ FULL EMPLOYEE DASHBOARD
# -----------------------------
print("\n--- 1Ô∏è‚É£2Ô∏è‚É£ Testing Full Employee Dashboard ---")
r = requests.get(f"{BASE_URL}/hrms/employee/{EMPLOYEE_ID}/full-details")
check(r, "Full dashboard")
print(r.json())


print("\n================================")
print("‚úÖ‚úÖ‚úÖ ALL EMPLOYEE API TESTS PASSED ‚úÖ‚úÖ‚úÖ")
print("================================\n")


```
