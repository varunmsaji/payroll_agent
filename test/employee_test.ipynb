```python
import requests

BASE = "http://localhost:8000/hrms/shifts"


# ------------------------------------------------------------
# 1Ô∏è‚É£ TEST: Get all shifts
# ------------------------------------------------------------
def test_get_all_shifts():
    print("\nüîπ GET /shifts")
    r = requests.get(f"{BASE}/")
    print("Status:", r.status_code)
    print(r.json())


# ------------------------------------------------------------
# 2Ô∏è‚É£ TEST: Create a shift
# ------------------------------------------------------------
def test_create_shift():
    print("\nüîπ POST /shifts")

    payload = {
        "shift_name": "Test Shift",
        "start_time": "10:00:00",
        "end_time": "18:00:00",
        "is_night_shift": False
    }

    r = requests.post(f"{BASE}/", json=payload)
    print("Status:", r.status_code)
    data = r.json()
    print(data)
    return data.get("shift_id")


# ------------------------------------------------------------
# 3Ô∏è‚É£ TEST: Get shift by ID
# ------------------------------------------------------------
def test_get_shift(shift_id):
    print(f"\nüîπ GET /shifts/{shift_id}")
    r = requests.get(f"{BASE}/{shift_id}")
    print("Status:", r.status_code)
    print(r.json())


# ------------------------------------------------------------
# 4Ô∏è‚É£ TEST: Update shift
# ------------------------------------------------------------
def test_update_shift(shift_id):
    print(f"\nüîπ PUT /shifts/{shift_id}")

    payload = {
        "shift_name": "Updated Test Shift",
        "start_time": "09:00:00",
        "end_time": "17:00:00",
        "is_night_shift": False
    }

    r = requests.put(f"{BASE}/{shift_id}", json=payload)
    print("Status:", r.status_code)
    print(r.json())


# ------------------------------------------------------------
# 5Ô∏è‚É£ TEST: Delete shift
# ------------------------------------------------------------
def test_delete_shift(shift_id):
    print(f"\nüîπ DELETE /shifts/{shift_id}")
    r = requests.delete(f"{BASE}/{shift_id}")
    print("Status:", r.status_code)
    print(r.json())


# ------------------------------------------------------------
# 6Ô∏è‚É£ TEST: Assign shift to employee
# ------------------------------------------------------------
def test_assign_shift(employee_id, shift_id):
    print("\nüîπ POST /shifts/assign")
    payload = {
        "employee_id": employee_id,
        "shift_id": shift_id,
        "effective_from": "2025-01-01"
    }
    r = requests.post(f"{BASE}/assign", json=payload)
    print("Status:", r.status_code)
    print(r.json())


# ------------------------------------------------------------
# 7Ô∏è‚É£ TEST: Employee shift history
# ------------------------------------------------------------
def test_shift_history(employee_id):
    print(f"\nüîπ GET /shifts/employee/{employee_id}/history")
    r = requests.get(f"{BASE}/employee/{employee_id}/history")
    print("Status:", r.status_code)
    print(r.json())


# ------------------------------------------------------------
# RUN ALL TESTS
# ------------------------------------------------------------
if __name__ == "__main__":
    # 1. Get existing shifts
    test_get_all_shifts()

    # 2. Create new shift
    new_shift_id = test_create_shift()

    if new_shift_id:
        # 3. Get newly created shift
        test_get_shift(new_shift_id)

        # 4. Update shift
        test_update_shift(new_shift_id)

        # 5. Assign shift to employee ID = 1
        test_assign_shift(employee_id=34, shift_id=new_shift_id)

        # 6. Get shift history for employee 1
        test_shift_history(employee_id=1)

        # 7. Finally delete the shift
        test_delete_shift(new_shift_id)
```

```python
import requests

# ============================
# CONFIG
# ============================

BASE_URL = "http://localhost:8000/hrms/leaves"
HEADERS = {"Content-Type": "application/json"}

EMPLOYEE_ID = 5     # change as needed
YEAR = 2025
MONTH = 2
MANAGER_ID = 101
LEAVE_ID_FOR_ACTION = 12  # for approve/reject testing


# ============================
# 1Ô∏è‚É£ DASHBOARD STATS
# ============================

def get_dashboard_stats():
    res = requests.get(f"{BASE_URL}/requests")
    data = res.json()

    stats = {
        "total": len(data),
        "pending": len([l for l in data if l["status"] == "pending"]),
        "approved": len([l for l in data if l["status"] == "approved"]),
        "rejected": len([l for l in data if l["status"] == "rejected"]),
    }

    print("\nüìä DASHBOARD STATS")
    print(stats)
    return stats


# ============================
# 2Ô∏è‚É£ MANAGER PENDING APPROVALS
# ============================

def get_pending_approvals():
    res = requests.get(f"{BASE_URL}/requests/pending")
    print("\n‚è≥ PENDING APPROVALS")
    print(res.json())
    return res.json()


# ============================
# 3Ô∏è‚É£ EMPLOYEE APPLIED LEAVES
# ============================

def get_employee_leaves(employee_id):
    res = requests.get(f"{BASE_URL}/requests/{employee_id}")
    print(f"\nüßë‚Äçüíº LEAVES APPLIED BY EMPLOYEE {employee_id}")
    print(res.json())
    return res.json()


# ============================
# 4Ô∏è‚É£ EMPLOYEE LEAVE BALANCE
# ============================

def get_employee_balance(employee_id, year):
    res = requests.get(f"{BASE_URL}/balance/{employee_id}/{year}")
    print(f"\nüìÅ LEAVE BALANCE | EMPLOYEE {employee_id} | YEAR {year}")
    print(res.json())
    return res.json()


# ============================
# 5Ô∏è‚É£ EMPLOYEE LEAVE HISTORY
# ============================

def get_employee_history(employee_id):
    res = requests.get(f"{BASE_URL}/history/{employee_id}")
    print(f"\nüïò LEAVE HISTORY | EMPLOYEE {employee_id}")
    print(res.json())
    return res.json()


# ============================
# 6Ô∏è‚É£ EMPLOYEE SALARY AFTER LEAVES
# ============================

def get_employee_salary(employee_id, year, month):
    res = requests.get(
        f"{BASE_URL}/salary/{employee_id}/{year}/{month}"
    )

    print(f"\nüí∞ SALARY AFTER LEAVES | EMPLOYEE {employee_id}")
    print(res.json())
    return res.json()


# ============================
# 7Ô∏è‚É£ APPROVE LEAVE
# ============================

def approve_leave(leave_id, manager_id):
    res = requests.post(
        f"{BASE_URL}/approve/{leave_id}",
        headers=HEADERS,
        json={"manager_id": manager_id}
    )

    print(f"\n‚úÖ APPROVE LEAVE | LEAVE ID {leave_id}")
    print(res.json())
    return res.json()


# ============================
# 8Ô∏è‚É£ REJECT LEAVE
# ============================

def reject_leave(leave_id, manager_id):
    res = requests.post(
        f"{BASE_URL}/reject/{leave_id}",
        headers=HEADERS,
        json={"manager_id": manager_id}
    )

    print(f"\n‚ùå REJECT LEAVE | LEAVE ID {leave_id}")
    print(res.json())
    return res.json()


# ============================
# üî• MAIN EXECUTION (FULL DASHBOARD FLOW)
# ============================

if __name__ == "__main__":

    print("\n============================")
    print("üöÄ LEAVES DASHBOARD TESTING")
    print("============================")

    # 1. Dashboard Stats
    get_dashboard_stats()

    # 2. Pending Approvals (Manager)
    get_pending_approvals()

    # 3. Employee Leaves
    get_employee_leaves(EMPLOYEE_ID)

    # 4. Employee Leave Balance
    get_employee_balance(EMPLOYEE_ID, YEAR)

    # 5. Employee Leave History
    get_employee_history(EMPLOYEE_ID)

    # 6. Employee Salary After Leaves
    get_employee_salary(EMPLOYEE_ID, YEAR, MONTH)

    # --------------------------------------------------
    # Uncomment ONLY ONE of these when testing actions
    # --------------------------------------------------

    # 7. Approve Leave
    # approve_leave(LEAVE_ID_FOR_ACTION, MANAGER_ID)

    # 8. Reject Leave
    # reject_leave(LEAVE_ID_FOR_ACTION, MANAGER_ID)

    print("\n‚úÖ DASHBOARD DATA FETCH COMPLETE")

```
