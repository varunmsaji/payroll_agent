```python
import requests

BASE_URL = "http://localhost:8000"   # FastAPI server
IMAGE_PATH = "test.jpg"

EMPLOYEE_ID = "26"   # ðŸ‘ˆ register face with ID 25


def register_face():
    url = f"{BASE_URL}/faces/register"

    with open(IMAGE_PATH, "rb") as img:
        files = {
            "image": (
                "test.jpg",
                img,
                "image/jpeg"
            )
        }

        data = {
            "employee_id": EMPLOYEE_ID
        }

        response = requests.post(url, files=files, data=data)

    print("STATUS:", response.status_code)

    try:
        print("RESPONSE:")
        print(response.json())
    except Exception:
        print(response.text)


if __name__ == "__main__":
    register_face()
```
